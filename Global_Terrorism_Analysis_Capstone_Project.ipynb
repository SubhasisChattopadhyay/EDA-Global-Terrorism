## <b> The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.</b>

# <b> Explore and analyze the data to discover key findings pertaining to terrorist activities. </b>

**Co-relation between No.of Terror attacks and No.of Terror Groups active in a country** 

```python
country_frequency=pd.DataFrame(df.country_txt.value_counts())
country_frequency
```

```python
#forming a dataset having country name, no.of terror organisation present in the country and no.of Terror incidents in the country
active_groups=df.groupby('country_txt')['gname'].nunique().reset_index()
active_groups=pd.merge(active_groups, country_frequency, right_on="index",left_on='country_txt')
active_groups.drop('index', axis=1, inplace=True)
active_groups
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x= "gname", y='country_txt_y', data= active_groups)
plt.title("Corelation between Terrorist attacks and Terrorist groups operates in a Country")
plt.xlabel("No.of Active Groups")
plt.ylabel('No.ofTerror Attacks')

```

```python
#genetating a histogram depicting no.of Terrorist Groups and No.of Terror attacks in most affected countries
plt.figure(figsize= (15,12))
x=np.arange(10)
plt.bar(x-0.2,atkVSgrps.gname,0.4,label='No.of Terrorist Groups')
plt.bar(x+0.2,atkVSgrps.country_txt_y,0.4,label='No.of Terrorist attacks')
plt.xticks(x,atkVSgrps.country_txt_x)
plt.legend()
plt.show()
```
